## 4. Data Engineering
<a class="anchor" id="1.1"></a>
<a href=#cont>Back to Table of Contents</a>

---
    
| ⚡ Description: Data engineering ⚡ |
| :--------------------------- |
| In this section you are required to: clean the dataset, and possibly create new features - as identified in the EDA phase. |

---

```python
pred_name='load_shortfall_3h'
var_names = list(Bilbao_data.columns)
var_names.remove(pred_name)
var_data = Bilbao_data[var_names]
pred_data = Bilbao_data[pred_name]

scaler = MinMaxScaler()
var_scaled = scaler.fit_transform(var_data)
var_normalize = pd.DataFrame(var_scaled, columns=var_data.columns)
var_normalize
# remove missing values/ features
```

```python
corrs = Bilbao_data.corr()['load_shortfall_3h'].sort_values(ascending=False)
corrs
# Build a dictionary of correlation coefficients and p-values
var_dict_cp = {}

column_titles = [col for col in corrs.index if col!= 'load_shortfall_3h']
for col in column_titles:
    p_val = round(pearsonr(Bilbao_data[col], Bilbao_data['load_shortfall_3h'])[1],8)
    var_dict_cp[col] = {'Correlation_Coefficient':corrs[col],
                    'P_Value':p_val}
    
var_df_cp = pd.DataFrame(var_dict_cp).T
var_df_cp_sorted = var_df_cp.sort_values('P_Value')
var_df_cp_sorted[var_df_cp_sorted['P_Value']<0.01]
# create new features
```

```python
pred_data = Bilbao_data[pred_name]  

# Model building - Independent Variable (IV) DataFrame
X_names = list(var_df_cp[var_df_cp['P_Value'] < 0.1].index)
X_data = Bilbao_data[X_names]# engineer existing features

corr = X_data.corr()

# Find rows and columnd where correlation coefficients > 0.9 or <-0.9
corr[np.abs(corr) > 0.9]

corr = X_data.corr()
r, c = np.where(np.abs(corr) > 0.9)

# We are only interested in the off diagonal entries:
off_diagonal = np.where(r != c)

# Show the correlation matrix rows and columns where we have highly correlated off diagonal entries:
corr.iloc[r[off_diagonal], c[off_diagonal]][0:3]
```
