```python
#Dummy Variable encoding
valencia_dummies = pd.get_dummies(valencia, drop_first=True)
column_index = [col for col in valencia_dummies.columns if col !='load_shortfall_3h' ] + ["load_shortfall_3h"]
valencia_dummies = valencia_dummies.reindex(columns=column_index)
valencia_dummies.head()
```

```python
#heatmap for the valencia variables
from statsmodels.graphics.correlation import plot_corr

fig = plt.figure(figsize=(15,15));
ax = fig.add_subplot(111);
plot_corr(valencia_dummies.corr(), xnames = valencia_dummies.corr().columns, ax = ax);
```

```python
from statsmodels.formula.api import ols
v_fm = valencia_dummies.copy()
y_name = 'load_shortfall_3h' # The dependent|Target variable
#using a for loop to check where the column name isn't equal to  'load_shortfall_3h' 
X_names = [col for col in valencia_dummies.columns if col != y_name]# The independent variable

# Build the OLS formula string " y ~ X "
formula_str = y_name+" ~ "+" + ".join(X_names);
model=ols(formula=formula_str, data=v_fm) #fit the model using the valencia dataframe
fitted = model.fit()
print(fitted.summary())

```

```python
#correlations between predictor variables and the target variable
corrs = valencia_dummies.corr()["load_shortfall_3h"].sort_values(ascending=False)

corrs
```

```python
from scipy.stats import pearsonr
my_dict = {} #buildig a dictionary of correlation coefficients and p_values
corrs = valencia_dummies.corr()["load_shortfall_3h"].sort_values(ascending=False)
column_titles = [col for col in corrs.index if col!="load_shortfall_3h"]
for col in column_titles:
    p_val = round(pearsonr(valencia_dummies[col], valencia_dummies['load_shortfall_3h'])[1],6)
    my_dict[col] = {'Correlation_Coefficient': corrs[col], 'P_Value': p_val}
val = pd.DataFrame(my_dict).T
val_sorted = val.sort_values("P_Value")
val_sorted[val_sorted['P_Value']<0.05]
```

Using a P_value of 0.1 and 0.05, both Valencia_wind_deg_level_7 and Valencia_wind_deg_level_6 both failed the test, being statistically insignificant and should be dropped.

Using the above info(heatmap and the val dataframe) the unfiltered set of independent variables would be retained for futuristic tests.

```python
y_data = valencia_dummies[y_name]# storing the target variable inside y_data
X_names = list(val[val["P_Value"]< 0.05].index)
X_data = valencia_dummies[X_names]
```

```python
X_data.shape
```

TEST FOR MULTICOLINEARITY 
correlation should only exist between individual independent variables and the target variables, to obtain an unbiased relationship during modelling.

```python
corr = X_data.corr()  #Storing the correlation coeff in a corr variable
r, c = np.where(np.abs(corr)>0.9)  #Finding rows and columns where correlaion coefficients >0.9 or <0.9
result = np.where(r != c) #Finding interest in the off_diagonal entries
corr.iloc[r[result], c[result]] # Showing correlation matrix where there's highly correlated result
```

Taking a look at the above DataFrame,It shows that there's a strong positive correlation among the Valencia_temp, Valencia_temp_max, and Valencia_temp_min Variables.
Although the Valencia_temp, Valencia_temp_max, and  Valencia_temp_min Variables all have the same p_value as seen from the val DataFrame; the Valencia_temp_max is slightly better correlated  to the dependent variable load_shortfall_3h,thereby having a higher predicting power to the target variable. Therefore, the Valencia_temp_max will be retained and the other two variables will be dropped off from the feature dataframe.

```python
# The resulting subset of the independent variable
X_omit = ['Valencia_temp','Valencia_temp_min'] 
X_corr = [col for col in X_names if col not in X_omit]

formula_str = y_name+' ~ '+' + '.join(X_corr); # Creating a new OLS formula from the smaller subset
print('Formula:\n\t{}'.format(formula_str))
model = ols(formula=formula_str,data= v_fm ) #Fit the model using th model dataframe
fitted = model.fit()
print(fitted.summary()) # display the fitted sumary
```

DATA SCALING USING NORMALIZATION

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
X_scaled = scaler.fit_transform(X_data)
X_normalize = pd.DataFrame(X_scaled, columns=X_data.columns)
```

```python
X_normalize.head()
```

```python
#X_normalize.describe().loc['std']
```

DATA SCALING(NORMAZATION) HAS BEEN APPLIED TO THE ABOVE DATASET

```python
from sklearn.feature_selection import VarianceThreshold
selector = VarianceThreshold(threshold=0.03) #Create a varianceThreshold object, selector
selector.fit(X_normalize)
```

```python
col_variances = selector.variances_

vars_dict = {}
vars_dict = [{"Variable_Name": c_name, "Variance": c_var}
             for c_name, c_var in zip(X_normalize.columns, col_variances)]
df_vars = pd.DataFrame(vars_dict)
df_vars.sort_values(by='Variance', ascending=False)
```

```python
#Create new column whose variance is above the selected threshold of 0.03
X_column = X_normalize[X_normalize.columns[selector.get_support(indices=True)]]

# Save variable names for later
X_var_names = X_column.columns

# View first few entries
X_column.head()
```

AFTER NORMALIZING THE DATASET, THE AMOUNT OF PREDICTOR VARIABLES DROPPED FROM 16 TO 11 AS A RESULT OF THE SELECTED THRESHOLD OF 0.03.
