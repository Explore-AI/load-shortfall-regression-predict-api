```python
#univariate analysis

#Histograms 
features_load_shortfall_3h = ['load_shortfall_3h']  # create a histogram for the y variable 
df[features_load_shortfall_3h].hist(figsize=(10,10), bins=10);

features_wind_speed = ['Madrid_wind_speed', 'Valencia_wind_speed', 'Bilbao_wind_speed'] # create a list of all numerical features(wind_speed)
df[features_wind_speed].hist(figsize=(10,10), bins=10);

features_rain_1h = ['Bilbao_rain_1h'] # create a list of all numerical features (rain_1h)
df[features_rain_1h].hist();

features_humidity = ['Madrid_humidity', 'Seville_humidity'] # create a list of all numerical features (humidity)
df[features_humidity].hist(figsize=(10,10));

features_clouds_all = ['Bilbao_clouds_all'] # create a list of all numerical features(clouds_all)
df[features_clouds_all].hist(figsize=(10,10));

features_temp_max = ['Madrid_temp_max', 'Bilbao_temp_max',] # create a list of all numerical features (temp_max)
df[features_temp_max].hist(figsize=(10,10), bins=10);

features_temp_min = ['Madrid_temp_min', 'Bilbao_temp_min', 'Seville_temp_min', 'Barcelona_temp_min'] # create a list of all numerical features (temp_min)
df[features_temp_min].hist(figsize=(10,10), bins=10);

features_temp = ['Madrid_temp', 'Bilbao_temp', 'Barcelona_temp']  # create a list of all numerical features (Barcelona_temp)
df[features_temp].hist(figsize=(10,10), bins=10);
```

```python
#categorical bar plot for wind
fig_dims = (10, 6)
fig, ax = plt.subplots(figsize=fig_dims)
sns.countplot(x = 'Valencia_wind_deg', data = df, palette="hls")
#plt.title("")
```

```python
#Density Plots
df[features_wind_speed].plot(kind='density', subplots=True, layout=(3, 2), sharex=False, figsize=(10, 10));
df[features_rain_1h].plot(kind='density', subplots=True, layout=(3, 2), sharex=False, figsize=(10, 10));
df[features_humidity].plot(kind='density', subplots=True, layout=(3, 2), sharex=False, figsize=(10, 10));
df[features_clouds_all].plot(kind='density', subplots=True, layout=(3, 2), sharex=False, figsize=(10, 10));
df[features_temp_max].plot(kind='density', subplots=True, layout=(3, 2), sharex=False, figsize=(10, 10));
df[features_temp_min].plot(kind='density', subplots=True, layout=(3, 2), sharex=False, figsize=(10, 10));
df[features_temp].plot(kind='density', subplots=True, layout=(3, 2), sharex=False, figsize=(10, 10));
df[features_load_shortfall_3h].plot(kind='density', subplots=True, layout=(3, 2), sharex=False, figsize=(10, 10));
```

```python
#Boxplots
sns.boxplot(x='load_shortfall_3h', data=df);
sns.boxplot(x='Madrid_wind_speed', data=df);
sns.boxplot(x='Valencia_wind_speed', data=df);
sns.boxplot(x='Bilbao_wind_speed', data=df);
sns.boxplot(x='Bilbao_rain_1h', data=df);
sns.boxplot(x='Madrid_humidity', data=df);
sns.boxplot(x='Seville_humidity', data=df);
sns.boxplot(x='Bilbao_clouds_all', data=df);
sns.boxplot(x='Madrid_temp_max', data=df);
sns.boxplot(x='Bilbao_temp_max', data=df);
sns.boxplot(x='Madrid_temp_min', data=df);
sns.boxplot(x='Bilbao_temp_min', data=df);
sns.boxplot(x='Seville_temp_min', data=df);
sns.boxplot(x='Barcelona_temp_min', data=df);
sns.boxplot(x='Madrid_temp', data=df);
sns.boxplot(x='Bilbao_temp', data=df);
sns.boxplot(x='Barcelona_temp', data=df);
```
